# Exploratory Data Analysis for finding pattern , correlation between features and outlier detection.

## Here we visualizing a each feature distribution through histogram also visualizing the count values and detecting outliers in each numerical features

### Univariate Analysis for input features and their distribution over the population.

```python
sns.histplot(data=df['age'], kde=True, color='red')
plt.title('Histogram of Age with KDE')
plt.xlabel('Age')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))  # Adjust size as needed.
plt.boxplot(df['age'])
plt.title('Boxplot of Age')
plt.xlabel('Variables')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
df['sex'].value_counts().plot(kind='bar')
plt.title('Count of Each Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
#1 - Male.
#0 - female.
```

```python
df['chest pain type'].value_counts().plot(kind='bar', color = "purple")
plt.title('Count of Each chest pain type')
plt.xlabel('chest pain type')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
#Encodes the type of chest pain experienced:
# 1 = Typical angina.
# 2 = Atypical angina.
# 3 = Non-anginal pain.
# 4 = Asymptomatic.
```

```python
df['sex'] = df['sex'].map({
    0: 'Female',
    1: 'Male',
})
sns.barplot( x = df['target'],y = df['age'] , hue = df['sex'] , data = df)         
plt.title("BarPlot Between Age Vs Having a Disease in male and female")
plt.grid()
plt.show()
```

```python
df['exercise angina'] = df['exercise angina'].map({
    0: 'Normal',
    1: 'heart problem',
})
sns.scatterplot( x = df['age'],y = df['cholesterol'],hue = df['exercise angina'],style=df['target'])
```

```python
sns.set(style="whitegrid", palette="coolwarm")

# Create correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("🔍 Correlation Heatmap of Heart Disease Features", fontsize=14, weight='bold')
plt.show()
```
